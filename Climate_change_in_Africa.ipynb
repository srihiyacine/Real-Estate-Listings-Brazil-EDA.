```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px
import plotly.graph_objects as go
from scipy.stats import norm
```

```python
data_path = "/content/Africa_climate_change.csv"
df = pd.read_csv(data_path)
```

```python
df['DATE'] = pd.to_datetime(df['DATE'], format='%Y%m%d %H%M%S')

# Convert the datetime format to the desired format "%Y-%m-%d"
df['DATE'] = df['DATE'].dt.strftime('%Y-%m-%d')
```

```python
df.head()
```

```python
df.info()
```

```python
df['TAVG'] = df['TAVG'].fillna(method='ffill')
```

```python
df.info()
```

```python

df_c = df[df["COUNTRY"]== "Cameroon"]
df_t = df[df["COUNTRY"]== "Tunisia"]


fig = px.line(title='Average Temperature Fluctuations in Tunisia and Cameroon')

# Add lines for Tunisia and Cameroon
fig.add_scatter(x=df_t['DATE'], y=df_t['TAVG'], mode='lines', name='Tunisia')
fig.add_scatter(x=df_c['DATE'], y=df_c['TAVG'], mode='lines', name='Cameroon')

# Update layout
fig.update_xaxes(title='DATE', tickangle=45)
fig.update_yaxes(title='Average Temperature')
fig.show()
```

```python

df_s = df[df["COUNTRY"]=="Senegal"]

# Parameters
mean = df_s["TAVG"].mean()
std =  df_s["TAVG"].std()

# Generate data for the normal distribution
x = np.linspace(mean - 3 * std, mean + 3 * std, 1000)
y = norm.pdf(x, mean, std)

normal_curve = pd.DataFrame({'Temperature': x, 'Density': y})

df_1980_2000 = df_s[(df_s['DATE'] >= '1980-01-01') & (df_s['DATE'] <= '2000-12-31')]
df_2000_2023 = df_s[(df_s['DATE'] >= '2000-01-01') & (df_s['DATE'] <= '2023-12-31')]

# Create histograms for each time period using Plotly Express
fig = px.histogram(df_1980_2000, x='TAVG', title='Temperature Distribution in Senegal (1980-2000)')
fig.add_histogram(x=df_2000_2023['TAVG'], name='Temperature Distribution in Senegal (2000-2023)')
# Add normal distribution trace
fig.add_trace(px.line(normal_curve, x='Temperature', y='Density').data[0])

# Show the histograms
fig.show()
```
